```python
import os

centrality = "BC"
dataset = 'CA-GrQc'

if centrality == "BC":
    centrality_word = "betweenness centrality"
elif centrality == "close":
    centrality_word = "closeness centrality"
elif centrality == "degree":
    centrality_word = "degree centrality"
elif centrality == "comm":
    centrality_word = "community structure"

# Initialize a list to store all saved_fit_configs for rd0
all_saved_fit_configs_rd0 = []

end_rounds = []

# Load all 50 saved_fit_configs for rd0
for i in range(50):
    filename_rd0 = f'../Sorting/{centrality}/{dataset}/rd0/saved_fit_config{i}.npy'
    saved_fit_config_rd0 = np.load(filename_rd0)
    
    # Check if corresponding file exists for rd1
    filename_rd1 = f'../Sorting/{centrality}/{dataset}/rd1/saved_fit_config{i}.npy'
    if os.path.exists(filename_rd1):
        saved_fit_config_rd1 = np.load(filename_rd1)
        
        # Add the rd1 fit values to rd0 fit values
        saved_fit_config_rd0 = np.concatenate((saved_fit_config_rd0, saved_fit_config_rd1), axis=None)

    filename_rd2 = f'../Sorting/{centrality}/{dataset}/rd2/saved_fit_config{i}.npy'
    if os.path.exists(filename_rd2):
        saved_fit_config_rd2 = np.load(filename_rd2)
        
        # Add the rd1 fit values to rd0 fit values
        saved_fit_config_rd0 = np.concatenate((saved_fit_config_rd0, saved_fit_config_rd2), axis=None)

    filename_rd3 = f'../Sorting/{centrality}/{dataset}/rd3/saved_fit_config{i}.npy'
    if os.path.exists(filename_rd3):
        saved_fit_config_rd3 = np.load(filename_rd3)
        
        # Add the rd1 fit values to rd0 fit values
        saved_fit_config_rd0 = np.concatenate((saved_fit_config_rd0, saved_fit_config_rd3), axis=None)

    filename_rd4 = f'../Sorting/{centrality}/{dataset}/rd4/saved_fit_config{i}.npy'
    if os.path.exists(filename_rd4):
        saved_fit_config_rd4 = np.load(filename_rd4)
        #print(filename_rd4)
        
        # Add the rd1 fit values to rd0 fit values
        saved_fit_config_rd0 = np.concatenate((saved_fit_config_rd0, saved_fit_config_rd4), axis=None)
    
    all_saved_fit_configs_rd0.append(saved_fit_config_rd0)

    end_rounds.append(len(saved_fit_config_rd0))

# Plot all the saved_fit_configs for rd0 with rd1 added if available
plt.figure(figsize=(14, 6))
for i, saved_fit_config_rd0 in enumerate(all_saved_fit_configs_rd0):
    plt.plot(saved_fit_config_rd0, label=f'Configuration {i}')
plt.xlabel('Generation')
plt.ylabel('Fitness Value')
plt.title(f'{dataset} - Successive Halving for Sorting by {centrality_word}')
#plt.legend()
plt.grid(True)
#plt.savefig(f'../Sorting/{centrality}/{dataset} SH Sorting {centrality}.png', bbox_inches='tight')
plt.show()
```

# Reading Excel

```python
import pandas as pd


centrality = "comm"
dataset = "Blogs"

# Specify the path to your Excel file
#file_path = f"../Sorting/Sorting SH {centrality}.xlsx"
file_path = '../Mutation/Mutation SH.xlsx'
#file_path = 'Basic SH.xlsx'

#dataset = "CA-GrQc"

# Read the Excel file
df = pd.read_excel(file_path, sheet_name=f'{dataset}')

# Display the DataFrame
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

variables = ["popn_init", "p_selection", "n_points", "mutation"]

# Determine the number of rows and columns for the subplots grid
num_vars = len(variables)
cols = 2  # Set the number of columns
rows = (num_vars + cols - 1) // cols  # Calculate the number of rows needed

fig, axes = plt.subplots(rows, cols, figsize=(cols*5, rows*5))
axes = axes.flatten()  # Flatten the 2D array of axes to 1D for easier indexing

for idx, variable in enumerate(variables):
    summary_df = df.groupby([f'{variable}', 'runs']).size().reset_index(name='count')
    scaling_factor = 20  # Adjust as needed
    summary_df['count_scaled'] = summary_df['count'] * scaling_factor
    summary_df[f'{variable}'] = pd.Categorical(summary_df[f'{variable}'])
    summary_df = summary_df[summary_df['count'] > 0]

    x_values = summary_df[f'{variable}'].unique()
    category_index_mapping = {category: i for i, category in enumerate(x_values)}
    summary_df['x_coordinate'] = summary_df[f'{variable}'].map(category_index_mapping)
    
    ax = axes[idx]
    ax.scatter(summary_df['x_coordinate'], summary_df['runs'], s=summary_df['count_scaled'], alpha=0.8)
    
    if variable == "popn_init":
        ax.set_xlabel('p')
        ax.set_ylabel('Generations')
        #ax.set_title('Distribution of p')
    
    elif variable == "p_selection":
        ax.set_xlabel('Parental Selection Method')
        ax.set_ylabel('Generations')
        #ax.set_title('Distribution of Parental Selection')
        
    elif variable == "n_points":
        ax.set_xlabel('c')
        ax.set_ylabel('Generations')
        #ax.set_title('Distribution of c')
    
    elif variable == "mutation":
        ax.set_xlabel('Mutation Rate')
        ax.set_ylabel('Generations')
        #ax.set_title('Distribution of mutation rate')
        
    elif variable == "e_selection":
        ax.set_xlabel('Environmental Selection Method')
        ax.set_ylabel('Generations')
        #ax.set_title('Distribution of Environmental Selection')
    
    ax.set_xticks(range(len(x_values)))
    ax.set_xticklabels(x_values)
    ax.grid(True)

# Remove any empty subplots
for ax in axes[num_vars:]:
    fig.delaxes(ax)
    
fig.suptitle(f'GA w/ UAM - {dataset}', fontsize=16)
plt.tight_layout()
plt.savefig(f'../Mutation/{dataset} - combined.png', bbox_inches='tight')
plt.show()
```
